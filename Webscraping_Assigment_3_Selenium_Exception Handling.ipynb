Details to be scraped are: "Brand Name", "Name of the Product", "Price", "Return/Exchange", "Expected Delivery", "Availability" and “Product URL”. In case, if any of the details are missing for any of the product then replace it by “-“.

```python
# scrape all product urls
product_urls = []
start=0
end=3
for page in range(start,end):#for loop for scrapping 3 page
    url=driver.find_elements(By.XPATH,'//a[@class = "a-link-normal s-underline-text s-underline-link-text s-link-style a-text-normal"]')       #scraping urls
    for i in url:
        product_urls.append(i.get_attribute("href"))                        #appending the urls in product_urls list
    nxt_button=driver.find_element(By.XPATH,'//a[@class="s-pagination-item s-pagination-next s-pagination-button s-pagination-separator"]')     #scraping the list of buttons from the page
    nxt_button.click()
    time.sleep(2)
```

```python
# Making empty list to scrape data
Brand =[]
Smartphone =[]
Colour =[]
Storage_Rom =[]
Primary_camera=[]
Secondary_camera=[]
Display_size=[]
Price =[]
```

```python
len(URL)
```

```python
from tqdm import tqdm
for i in tqdm(URL):
    driver.get(i)
    time.sleep(3)
    
    # Expanding specification table by clicking on read more button
    try:
        Read_more=driver.find_element(By.XPATH,'//button[@class="_2KpZ6l _1FH0tX"]')
        Read_more.click()
        
    except NoSuchElementException:
        print('NoSuchElementException Occur')
        pass
    
    # Extracting Brand Name via Xpath
    try:
        brand=driver.find_element(By.XPATH,'//span[@class="B_NuCI"]')
        Brand.append(brand.text.split()[0])
    
    except NoSuchElementException:
        Brand.append('-')
        
    # Extracting Smartphone Model via Xpath
    try:
        smartphone=driver.find_element(By.XPATH,'//div[@class="_3k-BhJ"][1]/table/tbody/tr[3]/td[2]/ul/li')
        Smartphone.append(smartphone.text)
    
    except NoSuchElementException:
        Smartphone.append('-')
        
    # Extracting Colour via Xpath
    try:
        colour=driver.find_element(By.XPATH,'//div[@class="_3k-BhJ"][1]/table/tbody/tr[4]/td[2]/ul/li')
        Colour.append(colour.text)
    
    except NoSuchElementException:
        Colour.append('-')
    
    # Extracting Storage Rom via Xpath
    try:
        Rom=driver.find_element(By.XPATH,'//div[@class="_3k-BhJ"][4]/table/tbody/tr[1]/td[2]/ul/li')
        Storage_Rom.append(Rom.text)
        
    except NoSuchElementException:
        Storage_Rom.append('Not Mention')
        
    # Extracting primary camera detail via Xpath
    try:
        primary_camera=driver.find_element(By.XPATH,'//div[@class="_3k-BhJ"][5]/table/tbody/tr[2]/td[2]/ul/li')
        Primary_camera.append(primary_camera.text)
    
    except NoSuchElementException:
        Primary_camera.append('Not Mention')
        
    # Extracting Secondary Camera detail via Xpath
    try:
        secondary_cam=driver.find_element(By.XPATH,'//div[@class="_3k-BhJ"][5]/table/tbody/tr[5]/td[2]/ul/li')
        Secondary_camera.append(secondary_cam.text)
    
    except NoSuchElementException:
        Secondary_camera.append('Not Mention')
    
    # Extracting Display size via Xpath
    try:
        display=driver.find_element(By.XPATH,'//div[@class="_3k-BhJ"][2]/table/tbody/tr[1]/td[2]/ul/li')
        Display_size.append(display.text)
    except NoSuchElementException:
        Display_size.append('-')
    
           
    #Extracting Price via X path
    try:
        price=driver.find_element(By.XPATH,'//div[@class="_30jeq3 _16Jk6d"]')
        Price.append(price.text)
    except NoSuchElementException:
        Price.append('Not Mention')
```

# 7. Write a program to scrap all the available details of best gaming laptops from digit.in.

```python
# Connecting to webdriver
driver=webdriver.Chrome(r"C:\Users\User\Downloads\chromedriver_win32 (1)\chromedriver.exe")
time.sleep(1)

# Opening url in chrome browser
url="https://www.digit.in/laptops/"
driver.get(url)
time.sleep(3)
```

```python
#click on the top 10 gaming laptops
gaming_laptop=driver.find_element(By.XPATH,"//ul[@class='list-unstyled sidebar-list']/li[10]")
gaming_laptop.click()
```

```python
#extracting titles
title_list=[]
for i in driver.find_elements(By.XPATH,"//div[@class='right-container']/div//a/h3"):
    title_list.append(i.text.replace("\n",''))
print(title_list[0:2])

#seller details
seller_list=[]
for i in driver.find_elements(By.XPATH,"//td[@class='smmerchant']"):
    seller_list.append(i.text.replace("\n",''))
print(seller_list[0:2])
```

```python
#extracting details:
details=[]
for i in driver.find_elements(By.XPATH,"//div[@class='value']"):
    details.append(i.text)
details[0:4]

#As we can see we are getting 4 details from this list i.e. 1) os 2) display 3) processor 4) Memory so we shall segregrate the data accordingly to each spec

#seperating the list contents based on indexing
os_list=[]
display_list=[]
processor_list=[]
memory_list=[]

count=0
while count<int(len(details)):
    os_list.append(details[count])
    display_list.append(details[count+1])
    processor_list.append(details[count+2])
    memory_list.append(details[count+3])
    count=count+4 

#check if the details are segregatted
print(os_list[0:2])
print(display_list[0:2])
print(processor_list[0:2])
print(memory_list[0:2])
```

```python
#specifications
specs_list=[]
for i in driver.find_elements(By.XPATH,"//div[@class='Spcs-details']"):
    specs_list.append(i.text.replace("\n","").split("SPECIFICATIONOS :"))
print(specs_list[0:2])

#overview
overview_list=[]
for i in driver.find_elements(By.XPATH,"//div[@class='Section-center']/p"):
    overview_list.append(i.text)
print(overview_list[-1])
```

```python
print(len(title_list))
print(len(seller_list))
print(len(os_list))
print(len(processor_list))
print(len(display_list))
print(len(memory_list))
print(len(specs_list))
```

```python
#creating dataframe
df=pd.DataFrame({})
df['Title']=title_list
df['Seller']=seller_list
df['OS']=os_list
df['Processor']=processor_list
df['Display']=display_list
df['Memory']=memory_list
df["Specifications"]=specs_list


df
```

```python
# Connecting to webdriver
driver=webdriver.Chrome(r"C:\Users\User\Downloads\chromedriver_win32 (1)\chromedriver.exe")
time.sleep(1)


driver.get("https://www.youtube.com/")
search_tag=driver.find_element(By.XPATH,'/html/body/ytd-app/div[1]/div/ytd-masthead/div[3]/div[2]/ytd-searchbox/form/div[1]/div[1]/input')
search_tag.send_keys('story of ronaldo')

search=driver.find_element(By.XPATH,'/html/body/ytd-app/div[1]/div/ytd-masthead/div[3]/div[2]/ytd-searchbox/button')
search.click() 
time.sleep(3)
```

```python
Peaches_video=driver.find_element(By.XPATH,'/html/body/ytd-app/div[1]/ytd-page-manager/ytd-search/div[1]/ytd-two-column-search-results-renderer/div/ytd-section-list-renderer/div[2]/ytd-item-section-renderer/div[3]/ytd-video-renderer[1]/div[1]/div/div[1]/div/h3/a')
Peaches_video.click()
```

```python
comments=[]
time_posteds=[]
Likes=[]
No_of_Likes=[]
```

```python
# 500 time we scroll down by 10000 in order to generate more Comments
for _ in range(500):
    driver.execute_script("window.scrollBy(0,500)")
```

```python
comment=driver.find_elements(By.ID,"content-text")
for i in comment[0:501]:
       comments.append(i.text)
```

```python
time_posted=driver.find_elements(By.XPATH,'//yt-formatted-string[@class="published-time-text style-scope ytd-comment-renderer"]')
for i in time_posted[0:501]:
       time_posteds.append(i.text)
```

```python
like = driver.find_elements(By.XPATH,"//span[@class='style-scope ytd-comment-action-buttons-renderer']")
for i in like:
    Likes.append(i.text)
    
for i in range(1,len(Likes),2)[0:501]:
    No_of_Likes.append(Likes[i])
```

```python
print(len(comments),len(time_posteds),len(No_of_Likes))
```

```python
df=pd.DataFrame({'Comments':comments,'Time_Posted':time_posteds,'No_of_Likes':No_of_Likes})
df
```

# 10. Write a python program to scrape a data for all available Hostels from https://www.hostelworld.com/ in “London” location. You have to scrape hostel name, distance from city centre, ratings, total reviews, overall reviews, privates from price, dorms from price, facilities and property description¶

```python
# Connecting to webdriver
driver=webdriver.Chrome(r'C:/chromedriver.exe')
time.sleep(1)

# Opening url in chrome browser
url="https://www.hostelworld.com"
driver.get(url)
time.sleep(2)
```

```python
#locating the location search bar
Search = driver.find_element(By.XPATH,'//*[@id="search-input-field"]')
# Sending input London in search bar
Search.send_keys("London")
time.sleep(1)
```

```python
#select london
london = driver.find_element(By.XPATH,'//*[@id="predicted-search-results"]/li[2]/div')
london.click()
time.sleep(2)
```

```python
# do click on search button
Search_button = driver.find_element(By.XPATH,'/html/body/div[3]/div/div/div[2]/div[1]/div/div/div[4]/div/div[2]/div/div[5]/button')
Search_button.click()
```

```python
#lets find required data
hostel_name = []
distance = []
rating = []
reviews = []
over_all = []
pvt_prices = []
dorms_price = []
facilities = []
description =[]
product_url = []
```

```python
for i in driver.find_elements(By.XPATH,"//div[@class = 'pagination-item pagination-current' or @class='pagination-item']"):
    i.click()
    time.sleep(4)
```

```python
#fetching hostel name
try:
        name = driver.find_elements(By.XPATH,'//h2[@class="title title-6"]')
        for i in name:
            hostel_name.append(i.text)
except NoSuchElementException:
        hostel_name.append('-')
```

```python
print(hostel_name)
len(hostel_name)
```

```python
#fetching distance from city centre
    
try:
        dist = driver.find_elements(By.XPATH,"//div[@class='subtitle body-3']//a//span[1]")
        for i in dist:
            distance.append(i.text.replace('Hostel - ',''))
except NoSuchElementException:
        distance.append('-')
```

```python
print(distance)
len(distance)
```

```python
#fetch privates from price
try:
    for i in driver.find_elements(By.XPATH,'//div[@class="prices-col"]/a[@class="prices"]/div[1]'):
        pvt_prices.append(i.text)
except NoSuchElementException:    
    pvt_prices.append('-') 
print(len(pvt_prices))
pvt_prices
```

```python
#fetching dorms from price
#fetch privates from price
try:
    for i in driver.find_elements(By.XPATH,'//div[@class="prices-col"]/a[@class="prices"]/div[2]'):
        dorms_price.append(i.text)
except NoSuchElementException:    
    dorms_price.append('-') 
print(len(dorms_price))
dorms_price
```

```python
try:
    for i in driver.find_elements(By.XPATH,'//div[@class="property-card"]/div[@class="property"]/div[1]/div[2]'):
        facilities.append(i.text.replace('\n',', '))
except NoSuchElementException:
        facilities.append('-')
len(facilities)
facilities
```

```python
# Ratings
try: 
    for i in driver.find_elements(By.XPATH,'//div[@class="bottom-rating"]/a/div/div[1]'):
        if "No" in i.text:
            rating.append("No Ratings")
        else:
            rating.append(i.text)
except NoSuchElementException:
    rating.append('-')
print(len(rating))
rating
```

```python
#fetching total reviews
        
try:
    for i in driver.find_elements(By.XPATH,'//div[@class="bottom-rating"]/a/div/div/div[2]'):
        reviews.append(i.text)
except NoSuchElementException:
        reviews.append('-')
print(len(reviews))
reviews
```

```python
#fetching overall reviews     
try:
    for i in driver.find_elements(By.XPATH,'//div[@class="bottom-rating"]/a/div/div/div[1]'):
        over_all.append(i.text)
except NoSuchElementException:
        over_all.append('-')
print(len(over_all))
over_all
```

```python
#fetching property description     
try:
    for i in driver.find_elements(By.XPATH,'//span[@class="description"]'):
        description.append(i.text)
except NoSuchElementException:
        description.append('-')
print(len(description))
description
```

```python
#Creating the DataFrame
df=pd.DataFrame({'hostel_name':hostel_name,'distance':distance,'rating':rating,'reviews':reviews,'over_all':over_all,'pvt_prices':pvt_prices,'dorms_price':dorms_price,'facilities':facilities,'description':description})
df
```
